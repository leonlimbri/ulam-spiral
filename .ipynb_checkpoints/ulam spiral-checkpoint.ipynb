## Ulam Spiral
Creating an array of size `n x n`, where `n` is the pixel size of the ulam spiral image. With `n=5`, we have ulam spiral as shown below, the highlighted yellow is a prime number.

#### Packages to be imported

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

#### Functions
Generate a function to create an array of spiralling numbers of size n (without denoting whether the number is prime or not).

```python
def move_dir(curr_direction = 'start'):
    '''
        Gives the next direction of the spiral movement,
        if no parameter given, it is assumed that the
        next direction is going to be 'r'.
        
        Parameters
        ----------
        curr_direction: str
            A string that is either 'u', 'd', 'l' or 'r',
            else it returns the next direction as 'r'.
            
        Returns
        -------
        tuple
            A tuple of size 3, where the value is the
            movement of the row, column and the direction
            it moves respectively.
        
    '''
    
    if   curr_direction == 'u': direction = 'l'
    elif curr_direction == 'd': direction = 'r'
    elif curr_direction == 'l': direction = 'd'
    elif curr_direction == 'r': direction = 'u'
    else: direction = 'r'
    
    if   direction == 'u': return (-1, 0, 'u')
    elif direction == 'd': return (1, 0, 'd')
    elif direction == 'l': return (0, -1, 'l')
    elif direction == 'r': return (0, 1, 'r')
    else: return (0, 0, 'err')
```

```python
def spiralling_numbers(n):
    '''
        Generate a spiralling number array of shape (n, n, 1)
        
        Parameters
        ----------
        n : int
            An integer determining the number of rows and columns
            
        Returns
        -------
        np.array
            A numpy array with shape of (n, n, 1) denoting the
            number for each of the sequence.
            
    '''
    
    ## Creates the array
    arr       = np.zeros((n, n, 1))
    
    ## Finding the starting positions.
    start_row = n/2 + 1 if n % 2 == 0 else (n + 1)/2
    start_col = n/2 if n % 2 == 0 else (n + 1)/2

    ## Iterating from the starting position to create the ulam spiral
    row       = int(start_row - 1)
    col       = int(start_col - 1)

    step_count       = 1
    step_counter     = 1
    inc_count        = 0
    
    row_adder        = move_dir()[0]
    col_adder        = move_dir()[1]
    curr_direction   = move_dir()[2]
    
    for i in range(1, n * n + 1):
        arr[row][col][0] += i
        if i == step_counter + step_count:
            row_adder      = move_dir(curr_direction)[0]
            col_adder      = move_dir(curr_direction)[1]
            curr_direction = move_dir(curr_direction)[2]
            step_counter = i
            if inc_count:
                step_count += 1
                inc_count  = 0
            else:
                inc_count += 1
        row += row_adder
        col += col_adder
        
    return arr
```

```python
def isPrime(number):
    '''
        Function to determine whether the number
        inserted is a prime number of not.
        
        Parameters
        ----------
        number: int
            A number to be determined.
            
        Returns
        -------
        bool
            A boolean, True means that the number is
            prime and False if it is not.
            
    '''
    
    if number <= 3:
        return number > 1
    if not number % 2 or not number % 3:
        return False
    i    = 5
    stop = int(number ** 0.5)
    
    while i <= stop:
        if not number % i or not number % (i + 2):
            return False
        i += 6
    
    return True
```

#### Create Ulam Sprial Images

```python
n          = 500
save_image = True
image_name = 'ulam_500.png'
```

```python
ulam = spiralling_numbers(n)
for r in range(ulam.shape[0]):
    for c in range(ulam.shape[1]):
        ulam[r][c][0] = 255 if isPrime(ulam[r][c][0]) else 0
```

```python
plt.imshow(ulam, cmap = 'gray')
plt.axis('off')
```

```python
if save_image: cv2.imwrite(image_name, ulam)
```
